```python
import cv2
from ultralytics import YOLO
import supervision as sv
import numpy as np

cap = cv2.VideoCapture(0)
model = YOLO("yolov8n-seg.pt")
while True:
    ret, frame = cap.read()
    results = model(frame, verbose=False)
    boxes = results[0].boxes.xyxy.tolist() #Рамка
    #print(boxes)
    frame=results[0].plot()
    classes = results[0].boxes.cls.tolist() #Класс
    #print(classes)
    confidences = results[0].boxes.conf.tolist()
    #print(confidences)
    if boxes!=[]:
        mask_imshow = results[0].masks[0].data.cpu().numpy().transpose(1, 2, 0)#Маска двухбитная 
        mask_imshow = mask_imshow.astype(np.uint8)
        contours, _ = cv2.findContours(mask_imshow, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE) 
        if len(contours) != 0: 
            maxc = max(contours, key=cv2.contourArea)
            rect = cv2.minAreaRect(maxc)
            box = cv2.boxPoints(rect)
            width = int(rect[1][0])
            height = int(rect[1][1])
            angle = int(rect[2])
            if width < height:
                angle = 90- angle
            else:
                angle = 90+(90-angle)
            cv2.drawContours(frame, maxc, -1, (0, 0, 255), 2)
            print(angle)
    cv2.imshow("yolov8", frame)
    cv2.imshow("mask_imshow", mask_imshow)

    if (cv2.waitKey(30) == 27):
        break
```
