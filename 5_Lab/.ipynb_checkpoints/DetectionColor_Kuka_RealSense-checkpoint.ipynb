```python
from kuka import Kuka
from openshowvar import *
import time
import Color_detection_RealSense
import numpy as np
import cv2
import pyrealsense2
from realsense_camera import *


def eql(num1, num2, err=10):
    if num1 < -1000 or num2 < -1000:
        return False
    return True if abs(num1 - num2) <= err else False


#Connect robot
robot = openshowvar(ip = '192.168.1.2', port = 7000)
kuka = Kuka(robot)

kuka.read_cartessian()
camera = RealsenseCamera()
a=200
b=200

def one():
    desired_point = np.array([y1, x1,-100, -180, 0, 0])
    kuka.ptp(desired_point) #Отправка координат в Kuka
    print('one')

def dno():
    desired_point2 = np.array([y1, x1,z1+10, -180, 0, 0])
    kuka.ptp(desired_point2) #Отправка координат в Kuka
    print('dno')

def two():
    desired_point1 = np.array([y1, x1, 0, -180, 0, 0])
    kuka.ptp(desired_point1) #Отправка координат в Kuka
    print('two')
    
def three():  
    global a
    global b
    desired_point1 = np.array([a, b, 0, -180, 0, 0])
    kuka.ptp(desired_point1) #Отправка координат в Kuka
    a=a-10
    print('three')
    
def five():
    desired_point1 = np.array([kuka.x_cartessian, kuka.y_cartessian, z1+(96.3-z1-10), -180, 0, 0])
    kuka.ptp(desired_point1) #Отправка координат в Kuka
    print('five')
    
def six():
    desired_point1 = np.array([kuka.x_cartessian, kuka.y_cartessian, 0, -180, 0, 0])
    kuka.ptp(desired_point1) #Отправка координат в Kuka
    print('six')    
      
def four():
    desired_point2 = np.array([-50, 0, -100, -180, 0, 0])
    kuka.ptp(desired_point2) #Отправка координат в Kuka

    
while True:    
    x1,y1,z1=class3_RealSense.coord(camera)
    time.sleep(2)
    x2,y2,z2=class3_RealSense.coord(camera) 
    print(z1,z2)
    if (eql(x1,x2)) and (eql(y1,y2)) and (eql(z1,z2)):
        one()
        while True:
            kuka.read_cartessian()
            if ((eql(kuka.x_cartessian,y1)) and (eql(kuka.y_cartessian,x1)) and (eql(kuka.z_cartessian,-100))):
                time.sleep(1)
                dno()
                kuka.open_grip()
                break
        while True:
            kuka.read_cartessian()
            if ((eql(kuka.x_cartessian,y1)) and (eql(kuka.y_cartessian,x1)) and (eql(kuka.z_cartessian,z1))):
                kuka.close_grip()
                break     
        while True:
            if (kuka.read_out(1)=='TRUE') and (kuka.read_out(2)=='FALSE'):
                two()
                break
        while True:
            kuka.read_cartessian()
            if ((eql(kuka.x_cartessian,y1)) and (eql(kuka.y_cartessian,x1)) and (eql(kuka.z_cartessian,0))):
                three()
                break
        while True:
            kuka.read_cartessian()
            if ((kuka.y_cartessian-b==0) and (eql(kuka.z_cartessian,0))):
                five()
                break        
        while True:
            kuka.read_cartessian()
            if ((kuka.y_cartessian-b==0) and (eql(kuka.z_cartessian,z1+(96.3-z1-10)))):
                kuka.open_grip()
                time.sleep(2)
                break
        while True:
            kuka.read_cartessian()
            if ((kuka.y_cartessian-b==0) and (eql(kuka.z_cartessian,z1+(96.3-z1-10)))):
                six()
                break        
        while True:
            kuka.read_cartessian()
            if ((kuka.y_cartessian-b==0) and (eql(kuka.z_cartessian,0))):
                four()
                time.sleep(3)
                break
                kuka.read_cartessian()
                if ((eql(kuka.x_cartessian,-50)) and (eql(kuka.y_cartessian,0)) and (eql(kuka.z_cartessian,-100))):     
                    break
    continue
```
