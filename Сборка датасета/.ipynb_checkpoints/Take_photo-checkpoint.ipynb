```python
import cv2
import numpy as np
import time

cap=cv2.VideoCapture(2)
cap.set(cv2.CAP_PROP_FRAME_WIDTH,1280)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT,720)

z=1
while z!=500:
    ret, frame=cap.read()
    time.sleep(1)
    cv2.imwrite(''+ str(z) + '.png', frame)
    z=z+1   
    if (cv2.waitKey(30) == 27):
        break
```
