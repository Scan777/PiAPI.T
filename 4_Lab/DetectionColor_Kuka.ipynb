```python
from kuka import Kuka
from openshowvar import *
import time
import Color_detection
import numpy as np
import cv2



def eql(num1, num2, err=10):
    if num1 < 0 or num2 < 0:
        return False
    return True if abs(num1 - num2) <= err else False


#Connect robot
robot = openshowvar(ip = '192.168.1.2', port = 7000)
kuka = Kuka(robot)

kuka.read_cartessian()
camera = cv2.VideoCapture(1)
a=100
b=200

def one():
    desired_point = np.array([y1, x1,0, -180, 0, 0])
    kuka.ptp(desired_point) #Отправка координат в Kuka
    print('one')

def dno():
    desired_point2 = np.array([y1, x1,72, -180, 0, 0])
    kuka.ptp(desired_point2) #Отправка координат в Kuka
    print('dno')

def two():
    desired_point1 = np.array([y1, x1, 0, -180, 0, 0])
    kuka.ptp(desired_point1) #Отправка координат в Kuka
    print('two')
    
def three():  
    global a
    global b
    desired_point1 = np.array([a, b, 0, -180, 0, 0])
    kuka.ptp(desired_point1) #Отправка координат в Kuka
    a=a-60
    print('three')
    
def five():
    desired_point1 = np.array([kuka.x_cartessian, kuka.y_cartessian, 72, -180, 0, 0])
    kuka.ptp(desired_point1) #Отправка координат в Kuka
    print('five')
    
def six():
    desired_point1 = np.array([kuka.x_cartessian, kuka.y_cartessian, 0, -180, 0, 0])
    kuka.ptp(desired_point1) #Отправка координат в Kuka
    print('six')    
      
def four():
    desired_point2 = np.array([-50, 0, 0, -180, 0, 0])
    kuka.ptp(desired_point2) #Отправка координат в Kuka

    
while True:    
    x1,y1,a2=Color_detect.coord(camera)
    time.sleep(2)
    x2,y2,a5=Color_detect.coord(camera) 
    if (eql(x1,x2)) and (eql(y1,y2)):
        one()
        print(a2,a5)
        while True:
            kuka.read_cartessian()
            if ((eql(kuka.x_cartessian,y1)) and (eql(kuka.y_cartessian,x1))):
                time.sleep(1)
                dno()
                kuka.open_grip()
                break
        while True:
            kuka.read_cartessian()
            if ((eql(kuka.x_cartessian,y1)) and (eql(kuka.y_cartessian,x1)) and (kuka.z_cartessian-(72)==0)):
                kuka.close_grip()
                break     
        while True:
            if (kuka.read_out(1)=='TRUE') and (kuka.read_out(2)=='FALSE'):
                two()
                break
        while True:
            kuka.read_cartessian()
            if ((eql(kuka.x_cartessian,y1)) and (eql(kuka.y_cartessian,x1)) and (eql(kuka.z_cartessian,0))):
                three()
                break
        while True:
            kuka.read_cartessian()
            if ((kuka.y_cartessian-b==0) and (eql(kuka.z_cartessian,0))):
                five()
                break        
        while True:
            kuka.read_cartessian()
            if ((kuka.y_cartessian-b==0) and (kuka.z_cartessian-(72)==0)):
                kuka.open_grip()
                time.sleep(2)
                break
        while True:
            kuka.read_cartessian()
            if ((kuka.y_cartessian-b==0) and (kuka.z_cartessian-(72)==0)):
                six()
                break        
        while True:
            kuka.read_cartessian()
            if ((kuka.y_cartessian-b==0) and (eql(kuka.z_cartessian,0))):
                four()
                time.sleep(3)
                break
                kuka.read_cartessian()
                if ((eql(kuka.x_cartessian,0)) and (eql(kuka.y_cartessian,0)) and (eql(kuka.z_cartessian,0))):     
                    break
    continue
```
